```python
import random
def birthday(iterations, people=23):
    k = 0
    for i in range(iterations):
        birthdays = [random.randint(1, 336) for g in range(people)]
        if len(birthdays) != len(set(birthdays)):
            k += 1
    denrozdeniya = (k / iterations) * 100
    return denrozdeniya
for u in range(10, 61):
    result = birthday(10000,u)
    print(f"Вероятность совпадения дней рождения в группе из {u} человек: {result:.2f}%")
```
